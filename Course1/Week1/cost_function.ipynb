Optional Lab 3: Cost Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_cost(x, y, w, b):
    """
    Computes the cost function for linear regression.

    Args:
      x (ndarray (m,)): Data, m examples
      y (ndarray (m,)): target values
      w,b (scalar)    : model parameters

    Returns
        total_cost (float): The cost of using w,b as the parameters for linear regression
               to fit the data points in x and y
    """
    m = x.shape[0]

    cost_sum = 0
    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum = cost_sum + cost
    total_cost = (1 / (2 * m)) * cost_sum

    return total_cost
```

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480, 430, 630, 730,])
```

```python
w_values = np.linspace(-1000, 1000, 1000)
b_values = np.linspace(-10, 10, 1000)

W, B = np.meshgrid(w_values, b_values)

costs = np.zeros(W.shape)

for i in range(len(w_values)):
    for j in range(len(b_values)):
        costs[j, i] = compute_cost(x_train, y_train, W[j, i], B[j, i])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, costs, cmap='viridis')
ax.set_xlabel('Weight w')
ax.set_ylabel('Bias b')
ax.set_zlabel('Cost')
ax.set_title('Cost Function Surface')
plt.show()
```

```python
min_cost_index = np.unravel_index(np.argmin(costs, axis=None), costs.shape)
w_min = W[min_cost_index]
b_min = B[min_cost_index]

print(f"Minimum cost is at w = {w_min:.2f}, b = {b_min:.2f}")
```
