```python
import matplotlib.pyplot as plt
import numpy as np
from lab_utils_multi import run_gradient_descent_feng, zscore_normalize_features
```

```python
x = np.array([2, 4, 6])
print("\nraw features:")
print(x)

X = np.c_[x, x**2, x**3]
print("\ncubic engineered features:")
print(X)

X = zscore_normalize_features(X)
print("\nzscore normalised features:")
print(X)
```

```python
# no feature engineering:
x = np.arange(0, 20, 1)
y = 1 + x**2
X = x.reshape(-1, 1)
model_w, model_b = run_gradient_descent_feng(X,y,iterations=1000, alpha = 1e-2)

plt.title("no feature engineering")
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.plot(x, X@model_w + model_b, label="Predicted Value")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# with x^2 feature engineering:
x = np.arange(0, 20, 1)
y = 1 + x**2
X = x**2
X = X.reshape(-1, 1)
model_w, model_b = run_gradient_descent_feng(X, y, iterations=10000, alpha = 1e-5)

plt.title("Added x**2 feature")
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.plot(x, np.dot(X,model_w) + model_b, label="Predicted Value")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# with x^3, x^2, x feature engineering:
x = np.arange(0, 20, 1)
y = x**2
X = np.c_[x, x**2, x**3]
model_w,model_b = run_gradient_descent_feng(X, y, iterations=10000, alpha=1e-7)

plt.title("x, x**2, x**3 features")
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.plot(x, X@model_w + model_b, label="Predicted Value")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
x = np.arange(0, 20, 1)
y = x**2
X = np.c_[x, x**2, x**3]
X_features = ['x','x^2','x^3']

fig, ax=plt.subplots(1, 3, figsize=(12, 3), sharey=True)
for i in range(len(ax)):
    ax[i].scatter(X[:,i],y)
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel("y")
plt.show()
```

```python
# feature engineering and normalisation
x = np.arange(0,20,1)
y = x**2
X = np.c_[x, x**2, x**3]
X = zscore_normalize_features(X)
model_w, model_b = run_gradient_descent_feng(X, y, iterations=100000, alpha=1e-1)

plt.title("Normalized x, x**2, x**3 feature")
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.plot(x, X@model_w + model_b, label="Predicted Value")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
